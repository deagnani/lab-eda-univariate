```python
import pandas as pd


data = pd.read_csv(r"C:\Users\deivi\Downloads\archive (1)\amz_uk_price_prediction_dataset.csv")
data.head()
```

```python

category_frequency = data['category'].value_counts()
category_frequency
```

```python
top_5_categories = category_frequency.head(5)
top_5_categories
```

```python
import matplotlib.pyplot as plt


top_categories = category_frequency.head(20)
top_categories.plot(kind='bar', figsize=(10, 5))
plt.title('Distribution of Top 20 Product Categories on Amazon UK')
plt.xlabel('Product Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

```python

top_5_categories.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Proportion of Top 5 Product Categories on Amazon UK')
plt.ylabel('')  
plt.show()
```

```python

mean_price = data['price'].mean()
median_price = data['price'].median()
mode_price = data['price'].mode()[0] 

print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")
```

```python
variance_price = data['price'].var()
std_dev_price = data['price'].std()
range_price = data['price'].max() - data['price'].min()
q1 = data['price'].quantile(0.25)
q3 = data['price'].quantile(0.75)
iqr = q3 - q1

print(f"Variance: {variance_price}")
print(f"Standard Deviation: {std_dev_price}")
print(f"Range: {range_price}")
print(f"Interquartile Range (IQR): {iqr}")
```

```python
plt.figure(figsize=(10, 5))
data['price'].hist(bins=50, edgecolor='black')
plt.title('Distribution of Product Prices on Amazon UK')
plt.xlabel('Price (£)')
plt.ylabel('Number of Products')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
data['price'].plot(kind='box')
plt.title('Box Plot of Product Prices on Amazon UK')
plt.ylabel('Price (£)')
plt.show()
```

```python

mean_stars = data['stars'].mean()
median_stars = data['stars'].median()
mode_stars = data['stars'].mode()[0] 
print(f"Mean Rating: {mean_stars}")
print(f"Median Rating: {median_stars}")
print(f"Mode Rating: {mode_stars}")
```

```python
variance_stars = data['stars'].var()
std_dev_stars = data['stars'].std()
q1_stars = data['stars'].quantile(0.25)
q3_stars = data['stars'].quantile(0.75)
iqr_stars = q3_stars - q1_stars

print(f"Variance: {variance_stars}")
print(f"Standard Deviation: {std_dev_stars}")
print(f"Interquartile Range (IQR): {iqr_stars}")
```

```python
skewness_stars = data['stars'].skew()
kurtosis_stars = data['stars'].kurt()

print(f"Skewness: {skewness_stars}")
print(f"Kurtosis: {kurtosis_stars}")
```

```python
plt.figure(figsize=(10, 5))
data['stars'].hist(bins=20, edgecolor='black')
plt.title('Distribution of Product Stars on Amazon UK')
plt.xlabel('Stars')
plt.ylabel('Number of Products')
plt.show()
```
